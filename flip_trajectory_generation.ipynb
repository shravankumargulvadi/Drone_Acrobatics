```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generateTrajectory(t,phi_1=180,phi_2=0,phi_3=180, W_max=1400):
    gamma_1=phi_1/W_max
    beta_1=-3/4*(gamma_1**-3)*W_max
    gamma_3=phi_3/W_max
    beta_3=-3/4*(gamma_3**-3)*W_max
    delta_1=2*gamma_1
    delta_2_prime=delta_1+phi_2/W_max
    if(t<=delta_1):
        W=beta_1/3*((t-gamma_1)**3)-beta_1*(gamma_1**2)*t+beta_1*(gamma_1**3)/3
    elif(t>delta_1 and t<=delta_2_prime):
        W=W_max
    else:
        W=beta_3/3*((gamma_3+delta_2_prime-t)**3)-beta_3*(gamma_3**2)*(2*gamma_3+delta_2_prime-t)+beta_3*(gamma_3**3)/3
    return W
   

```

## Single Flip

```python
t=np.linspace(0,0.5,100)
waypoints=[]
for t_i in t:
    waypoint=generateTrajectory(t_i)
    waypoints.append(waypoint)
plt.plot(t,waypoints)
print(waypoints)
```

## Double Flip

```python
t=np.linspace(0,0.7,100)
waypoints=[]
for t_i in t:
    waypoint=generateTrajectory(t_i,200,300,220,1600)
    waypoints.append(waypoint)
plt.plot(t,waypoints)
```

## Triple Flip

```python
t=np.linspace(0,0.85,100)
waypoints=[]
for t_i in t:
    waypoint=generateTrajectory(t_i,220,520,340,1900)
    waypoints.append(waypoint)
plt.plot(t,waypoints)
```

```python
import time
#t=np.linspace(0,0.5,100)
T=0.7
t_start=time.time()
t_current=time.time()
waypoints=[]
t_array=[]
t=abs(t_current-t_start)

while (t)<T:
    #waypoint=generateTrajectory(t)
    waypoint=generateTrajectory(t,200,300,220,1600)
    waypoints.append(waypoint)
    t_array.append(t)
    t_current=time.time()
    t=abs(t_current-t_start)
plt.plot(t_array,waypoints)
```
